```python
import datetime
import requests
import time
import pandas as pd
import numpy as np
import seaborn as sns
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)

# Test this strategy with longer backtesting period
# https://www.newtraderu.com/2020/08/13/a-monthly-trading-strategy-that-crushes-buy-and-hold/
# On the last day of the month if you are long the SPY ETF and price is going to close over the 
# 200-day simple moving average you stay long. 
# On the last day of the month if you are long the SPY ETF and price is going to close under the 
# 200-day simple moving average you sell your position and go to cash. 
# On the last day of the month if you in a cash position and price is going to close over the 
# 200-day simple moving average you buy the SPY ETF and go long. 
# On the last day of the month if you in a cash position and price is going to close under the 
# 200-day simple moving average you stay in cash. 

# Conclusion bad strategy.
# Conclusion tested various 50-day, 100-day SMA. They all perform very bad. 
# Conclusion 200-day SMA is a great signal to get into the market. We can get into the market earlier than that SMA signal, 
# but not too late after the SMA signal cross over if all the fundamentals are pointing towards better future

def getData():
    data = pd.read_csv("/Users/qiuchen/Documents/Documents/MyQuantFund/MomentumStrategy/SPY_daily.csv")
    data['Date'] = data.apply(lambda x: datetime.datetime.strptime(x['Date'], "%Y-%m-%d"), axis = 1)
    return data

def getBenchMarkStrategyPerformance(data, initialFund = 100, tax = True):
    initalDate = data.iloc[0]['Date']
    shares = initialFund / data.iloc[0]['Adj Close']
    res = []
    for index, row in data.iterrows():
        today = row['Date']
        cashAmount = shares * row['Adj Close'] 
        if tax:
            if cashAmount > initialFund: # profitable 
                if (today - initalDate).days > 365:
                    cashAmount = initialFund + (cashAmount - initialFund) * (0.8 - 0.038)
                else:
                    cashAmount = initialFund + (cashAmount - initialFund) * (1 - 0.222 - 0.066 - 0.037- 0.038)
            else:
                cashAmount = cashAmount
        res.append([today, cashAmount])
    return pd.DataFrame(data = res, columns = ['date', 'mv'])
    
def end_of_month(dt):
    todays_month = dt.month
    tomorrows_month = (dt + datetime.timedelta(days=1)).month
    return tomorrows_month != todays_month    

def applyStrategy(data):
    decision = [[data['Date'].tolist()[0], "cash", np.nan, np.nan]]
    for index, row in data.iterrows():
        if index < 200:
            continue
        else:
            sma = np.array(data[index-200:index]['Adj Close'].tolist()).mean()
            today = row['Date']
            if end_of_month(today):
                if decision[-1][1] == "cash":
                    if row['Adj Close'] >= sma:
                        decision.append([today, 'long', row['Adj Close'], sma])
                    else:
                        decision.append([today, 'cash', row['Adj Close'], sma])
                else:
                    if row['Adj Close'] >= sma:
                        decision.append([today, 'long', row['Adj Close'], sma])
                    else:
                        decision.append([today, 'cash', row['Adj Close'], sma])
    return pd.DataFrame(data = decision, columns = ['date', 'decision', 'adjclose', 'sma'])

def getPerformance(decision, initialFund = 100, tax = True):
    position = [[decision['date'].tolist()[0], 0, initialFund, np.nan, np.nan, 100, np.nan]]
    for index, row in decision.iterrows():
        if index == 0:
            continue
        if decision.iloc[index-1]['decision'] == 'cash' and row['decision'] == 'long':
            cashAmount = position[-1][2] #cash
            shares = cashAmount / row['adjclose'] 
            stockMV = cashAmount
            position.append([row['date'], shares, 0, row['adjclose'], stockMV, stockMV, 'buy'])
        elif decision.iloc[index-1]['decision'] == 'long' and row['decision'] == 'cash':
            cashAmount = position[-1][1] * row['adjclose'] 
            if tax:
                if cashAmount > position[-1][4]: # profitable 
                    if (row['date'] - position[-1][0]).days > 365:
                        cashAmount = position[-1][4] + (cashAmount - position[-1][4]) * (0.8 - 0.038)
                    else:
                        cashAmount = position[-1][4] + (cashAmount - position[-1][4]) * (1 - 0.222 - 0.066 - 0.037- 0.038)
                else:
                    cashAmount = cashAmount
            position.append([row['date'], 0, cashAmount, row['adjclose'], 0, cashAmount, 'sell'])
    return pd.DataFrame(data = position, columns = ['date', 'shares', 'cash', 'share_price', 'stockMV', 'mv', 'action'])
        
if __name__ == "__main__":
    startDate = datetime.date(1993,1,1)
    data = getData()
    data = data[data['Date'] >= pd.to_datetime(startDate)].reset_index(drop = True)
    decision = applyStrategy(data)
    performance = getPerformance(decision, tax = True).rename(columns = {'mv': 'mv_Strategy'})
    benchmarkPerformance = getBenchMarkStrategyPerformance(data, tax = True).rename(columns = {'mv': 'mv_Benchmark'})

    plotPd = benchmarkPerformance[['date', 'mv_Benchmark']].merge(performance[['date', 'mv_Strategy']], on = "date", how = "left")
    plotPd = plotPd.set_index('date')
    #plotPd = plotPd[plotPd.index >= pd.to_datetime(datetime.date(2016,1,1))]
    sns.set(rc={'figure.figsize':(11.7,8.27)})
    sns.lineplot(plotPd)
    
    
```
